## Verse of the Day

```python
# I want to read the output from the following RSS Link https://www.biblegateway.com/votd/get/?format=atom
# get the RSS feed
import feedparser
from html import unescape

# get the RSS feed
rss = feedparser.parse('https://www.biblegateway.com/votd/get/?format=atom')
# get value content
content = rss['entries'][0]['content'][0]['value']
# get the title
title = rss['entries'][0]['title']
# remove the &#8212; from the content
content = content.replace('&#8212;','; ')
# remove the &ldquo; and &rdquo; from the content
content = content.replace('&ldquo;','')
content = content.replace('&rdquo;','')

# Use the html.unescape method to convert HTML character references to their actual characters
content = unescape(content)

# print the content
print(content)
# print the title
print(title)
```

```python
## NEW MAIN CODE TO RUN

import requests
import json

code = 'app code'
redirect_uri = 'http://localhost'
client_id = 'client id'
client_secret = 'client secret'
AppID = 'app id'
AppKey = 'app key'

# Load previous tokens if they exist
try:
    with open('tokens.json', 'r') as f:
        tokens = json.load(f)
        access_token = tokens['access_token']
        refresh_token = tokens['refresh_token']
except FileNotFoundError:
    tokens = {}

# If access_token is not available, get it using authorization code
if 'access_token' not in tokens:
    url = "https://www.inoreader.com/oauth2/token"
    payload = {
        'code': code,
        'redirect_uri': redirect_uri,
        'client_id': client_id,
        'client_secret': client_secret,
        'scope': '',
        'grant_type': 'authorization_code'
    }
    headers = {
        'Content-Type': 'application/x-www-form-urlencoded'
    }
    response = requests.post(url, data=payload, headers=headers)
    if response.status_code == 200:
        data = response.json()
        access_token = data['access_token']
        refresh_token = data['refresh_token']
        with open('tokens.json', 'w') as f:
            json.dump(data, f)
    else:
        print('Error:', response.content)

# Function to refresh the access token
def refresh_access_token(refresh_token):
    url = "https://www.inoreader.com/oauth2/token"
    payload = {
        'refresh_token': refresh_token,
        'client_id': client_id,
        'client_secret': client_secret,
        'grant_type': 'refresh_token'
    }
    headers = {
        'Content-Type': 'application/x-www-form-urlencoded'
    }
    response = requests.post(url, data=payload, headers=headers)
    if response.status_code == 200:
        data = response.json()
        access_token = data['access_token']
        refresh_token = data['refresh_token']
        with open('tokens.json', 'w') as f:
            json.dump(data, f)
        return access_token, refresh_token
    else:
        print('Error:', response.content)

# You can call the refresh function when needed
# For example, if you receive a 401 status code (Unauthorized) when making a request to the API, it means your access token has expired and you should refresh it.
# access_token, refresh_token = refresh_access_token(refresh_token)

refresh_access_token(refresh_token)






import requests
import time
from collections import namedtuple

url = "https://www.inoreader.com/reader/api/0/stream/contents/user/-/label/Huskers"

headers = {
    'Authorization': 'Bearer ' + access_token,
    'AppID': AppID,
}

current_time = int(time.time())
start_time = current_time - 24*60*60  # 24 hours ago
article_details = []
seen_titles = set()  # This will keep track of titles we've seen

for timestamp in range(current_time, start_time, -8*60*60):  # Every hour back to 24 hours ago
    params = {
        'n': 100,
        'r': 'o',  # Return oldest first
        'ot': timestamp
    }
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        datum = response.json()
        items = datum['items']
        for item in items:
            if item['title'] in seen_titles:  # Skip this item if we've seen this title before
                continue
            seen_titles.add(item['title'])  # Add this title to our set of seen titles
            article_details.append({
                'id': item['id'],
                'categories': item['categories'],
                'title': item['title'],
                'published': item['published'],
                'updated': item['updated'],
                'canonical': item['canonical'][0]['href'],
                'summary_direction': item['summary']['direction'],
                'summary_content': item['summary']['content'],
                'author': item['author'],
                'origin_streamId': item['origin']['streamId'],
                'origin_title': item['origin']['title']
            })
    else:
        print('Error:', response.content)

# Define a namedtuple class 'Article' with the dictionary keys as fields
Article = namedtuple('Article', article_details[0].keys())

# Convert list of dictionaries to list of namedtuples
article_namedtuples = [Article(**detail) for detail in article_details]





# INITIALIZE THE LISTS
youtube_videos = []
articles = []

# Process each article
for article in article_details:
    article_url = article['canonical']

    if 'youtube' in article_url.lower():
        # Add YouTube videos to the youtube_videos list
        youtube_videos.append(article)
    else:
        # Add other articles to the articles list
        articles.append(article)



import logging
from newspaper import Article

# Configure logging
logging.basicConfig(filename='error.log', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# List to store the article texts
article_texts = []

# Loop through the article details and extract the article text
for article in articles:
    try:
        article_url = article['canonical']

        # Create a new Article object
        article_obj = Article(article_url)

        # Download and parse the article
        article_obj.download()
        article_obj.parse()

        # Get the article text
        article_text = article_obj.text

        # Append the article text to the list
        article_texts.append(article_text)
    except Exception as e:
        # Log the error and the URL that caused it
        logging.error(f"Error processing article with URL {article_url}: {e}")
        continue



import tiktoken
# count the number of tokens in a string
def count_tokens(string: str, encoding_name: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens

cost = 0
for text in article_texts:
    tokens = count_tokens(text, "cl100k_base")
    this_cost = tokens*0.003/1000
    cost += this_cost



def split_into_token_chunks(text, chunk_token_limit):
    """
    Split the text into chunks, each containing a specific number of tokens.

    Args:
        text (str): The text to split.
        chunk_token_limit (int): The maximum number of tokens for each chunk.

    Returns:
        list of str: The chunks of text.
    """
    words = text.split()
    chunks = []
    chunk = []
    token_count = 0

    for word in words:
        new_token_count = count_tokens(' '.join(chunk + [word]), "cl100k_base")

        if new_token_count <= chunk_token_limit:
            chunk.append(word)
            token_count = new_token_count
        else:
            chunks.append(' '.join(chunk))
            chunk = [word]
            token_count = count_tokens(word, "cl100k_base")

    if chunk:  # Add the last chunk if it's non-empty
        chunks.append(' '.join(chunk))

    return chunks




# Combine all articles into one big text
all_articles = ' '.join(article_texts)

# Split the combined text into chunks
token_limit_per_chunk = 12000  # Or any other value you prefer
chunks = split_into_token_chunks(all_articles, token_limit_per_chunk)



import openai

openai.api_key = "key"

system_message = {"role": "system", "content": "You assist a busy man in getting a daily news summary, but these particular stories are of extreme importance to him, as they provide talking points for his daily life. Make your summary detailed."}
outputs = []

for chunk in chunks:
    # Create messages
    messages = [system_message, {"role": "user", "content": "Here is some of today's news: " + chunk}]
    
    # Call the API
    response = openai.ChatCompletion.create(model="gpt-3.5-turbo-16k", messages=messages)
    
    # Save the response
    outputs.append(response.choices[0].message['content'])




# Combine all outputs into one text
all_outputs = ' '.join(outputs)

# Split the combined outputs into chunks
token_limit_per_chunk = 12000  # Adjust based on your needs
output_chunks = split_into_token_chunks(all_outputs, token_limit_per_chunk)




# Initialize the list of summaries
summaries = []

for chunk in output_chunks:
    # Create messages
    messages = [system_message, {"role": "user", "content": "You assist a busy man in getting a daily news summary, but these particular stories are of extreme importance to him, as they provide talking points for his daily life. Make your summary detailed: " + chunk}]

    # Call the API
    response = openai.ChatCompletion.create(model="gpt-3.5-turbo-16k", messages=messages)
    
    # Save the summary
    summaries.append(response.choices[0].message['content'])



# Combine all summaries into one text
final_summary = ' '.join(summaries)

# Print the final summary
print(final_summary)
```

```python
## NEW MAIN CODE TO RUN

import requests
import json

code = 'code'
redirect_uri = 'http://localhost'
client_id = 'id'
client_secret = 'secret'
AppID = 'id'
AppKey = 'key'

# Load previous tokens if they exist
try:
    with open('tokens.json', 'r') as f:
        tokens = json.load(f)
        access_token = tokens['access_token']
        refresh_token = tokens['refresh_token']
except FileNotFoundError:
    tokens = {}

# If access_token is not available, get it using authorization code
if 'access_token' not in tokens:
    url = "https://www.inoreader.com/oauth2/token"
    payload = {
        'code': code,
        'redirect_uri': redirect_uri,
        'client_id': client_id,
        'client_secret': client_secret,
        'scope': '',
        'grant_type': 'authorization_code'
    }
    headers = {
        'Content-Type': 'application/x-www-form-urlencoded'
    }
    response = requests.post(url, data=payload, headers=headers)
    if response.status_code == 200:
        data = response.json()
        access_token = data['access_token']
        refresh_token = data['refresh_token']
        with open('tokens.json', 'w') as f:
            json.dump(data, f)
    else:
        print('Error:', response.content)

# Function to refresh the access token
def refresh_access_token(refresh_token):
    url = "https://www.inoreader.com/oauth2/token"
    payload = {
        'refresh_token': refresh_token,
        'client_id': client_id,
        'client_secret': client_secret,
        'grant_type': 'refresh_token'
    }
    headers = {
        'Content-Type': 'application/x-www-form-urlencoded'
    }
    response = requests.post(url, data=payload, headers=headers)
    if response.status_code == 200:
        data = response.json()
        access_token = data['access_token']
        refresh_token = data['refresh_token']
        with open('tokens.json', 'w') as f:
            json.dump(data, f)
        return access_token, refresh_token
    else:
        print('Error:', response.content)

# You can call the refresh function when needed
# For example, if you receive a 401 status code (Unauthorized) when making a request to the API, it means your access token has expired and you should refresh it.
# access_token, refresh_token = refresh_access_token(refresh_token)

refresh_access_token(refresh_token)






import requests
import time
from collections import namedtuple

url = "https://www.inoreader.com/reader/api/0/stream/contents/1000003692"

headers = {
    'Authorization': 'Bearer ' + access_token,
    'AppId': AppID,
}

current_time = int(time.time())
start_time = current_time - 24*60*60  # 24 hours ago
article_details = []
seen_titles = set()  # This will keep track of titles we've seen

for timestamp in range(current_time, start_time, -60*60):  # Every hour back to 24 hours ago
    params = {
        'n': 100,
        'r': 'o',  # Return oldest first
        'ot': timestamp
    }
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        datum = response.json()
        items = datum['items']
        for item in items:
            if item['title'] in seen_titles:  # Skip this item if we've seen this title before
                continue
            seen_titles.add(item['title'])  # Add this title to our set of seen titles
            article_details.append({
                'id': item['id'],
                'categories': item['categories'],
                'title': item['title'],
                'published': item['published'],
                'updated': item['updated'],
                'canonical': item['canonical'][0]['href'],
                'summary_direction': item['summary']['direction'],
                'summary_content': item['summary']['content'],
                'author': item['author'],
                'origin_streamId': item['origin']['streamId'],
                'origin_title': item['origin']['title']
            })
    else:
        print('Error:', response.content)

# Define a namedtuple class 'Article' with the dictionary keys as fields
Article = namedtuple('Article', article_details[0].keys())

# Convert list of dictionaries to list of namedtuples
article_namedtuples = [Article(**detail) for detail in article_details]




# Use the current time as the initial timestamp
ts = int(time.time())
# This URL is used to mark all articles as read
url_mark_read = "https://www.inoreader.com/reader/api/0/mark-all-as-read"
headers = {
    'Authorization': 'Bearer ' + access_token
}
# Mark all articles as read
response = requests.get(url_mark_read, headers=headers, params={'s': '[streamId]', 'ts': ts})





# INITIALIZE THE LISTS
youtube_videos = []
articles = []

# Process each article
for article in article_details:
    article_url = article['canonical']

    if 'youtube' in article_url.lower():
        # Add YouTube videos to the youtube_videos list
        youtube_videos.append(article)
    else:
        # Add other articles to the articles list
        articles.append(article)



import logging
from newspaper import Article

# Configure logging
logging.basicConfig(filename='error.log', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# List to store the article texts
article_texts = []

# Loop through the article details and extract the article text
for article in articles:
    try:
        article_url = article['canonical']

        # Create a new Article object
        article_obj = Article(article_url)

        # Download and parse the article
        article_obj.download()
        article_obj.parse()

        # Get the article text
        article_text = article_obj.text

        # Append the article text to the list
        article_texts.append(article_text)
    except Exception as e:
        # Log the error and the URL that caused it
        logging.error(f"Error processing article with URL {article_url}: {e}")
        continue



import tiktoken
# count the number of tokens in a string
def count_tokens(string: str, encoding_name: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens

cost = 0
for text in article_texts:
    tokens = count_tokens(text, "cl100k_base")
    this_cost = tokens*0.003/1000
    cost += this_cost



def split_into_token_chunks(text, chunk_token_limit):
    """
    Split the text into chunks, each containing a specific number of tokens.

    Args:
        text (str): The text to split.
        chunk_token_limit (int): The maximum number of tokens for each chunk.

    Returns:
        list of str: The chunks of text.
    """
    words = text.split()
    chunks = []
    chunk = []
    token_count = 0

    for word in words:
        new_token_count = count_tokens(' '.join(chunk + [word]), "cl100k_base")

        if new_token_count <= chunk_token_limit:
            chunk.append(word)
            token_count = new_token_count
        else:
            chunks.append(' '.join(chunk))
            chunk = [word]
            token_count = count_tokens(word, "cl100k_base")

    if chunk:  # Add the last chunk if it's non-empty
        chunks.append(' '.join(chunk))

    return chunks




# Combine all articles into one big text
all_articles = ' '.join(article_texts)

# Split the combined text into chunks
token_limit_per_chunk = 12000  # Or any other value you prefer
chunks = split_into_token_chunks(all_articles, token_limit_per_chunk)



import openai

openai.api_key = "key"

system_message = {"role": "system", "content": "You assist a busy man in getting a daily news summary. Keep your summary of the news brief and readable. Most of the news is irrelevant, focus on the most important news. Feel free to sacrifice some details for readability and quantity of news covered."}
outputs = []

for chunk in chunks:
    # Create messages
    messages = [system_message, {"role": "user", "content": "Here is some of today's news: " + chunk}]
    
    # Call the API
    response = openai.ChatCompletion.create(model="gpt-3.5-turbo-16k", messages=messages)
    
    # Save the response
    outputs.append(response.choices[0].message['content'])




# Combine all outputs into one text
all_outputs = ' '.join(outputs)

# Split the combined outputs into chunks
token_limit_per_chunk = 12000  # Adjust based on your needs
output_chunks = split_into_token_chunks(all_outputs, token_limit_per_chunk)




# Initialize the list of summaries
summaries = []

for chunk in output_chunks:
    # Create messages
    messages = [system_message, {"role": "user", "content": "Please provide a summary of this text: " + chunk}]

    # Call the API
    response = openai.ChatCompletion.create(model="gpt-3.5-turbo-16k", messages=messages)
    
    # Save the summary
    summaries.append(response.choices[0].message['content'])



# Combine all summaries into one text
final_summary = ' '.join(summaries)

# Print the final summary
print(final_summary)
```
